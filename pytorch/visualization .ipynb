```python
import os, util
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt

def plot (logs, xlabel= "Epochs", ylabel='Loss'):
    keys = logs.keys()[1:]
    plt.figure()
    for k in keys:
        # normalize so that visulization is readable.
        plt.plot(logs[k], label=k)
    plt.legend()
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

### MLP - MNIST

```python
file_name = 'results/mlp/mlp_mnist_test_loss.csv'
file = pd.read_csv (file_name)
plot(file)
```

### ResNet18 - CIFAR10

```python
file_name = 'results/resnet18/resnet18_cifar10_test_loss.csv'
file = pd.read_csv (file_name)
plot(file)
```
